```python
import requests
import re
from requests.packages import urllib3
urllib3.disable_warnings()
headers={
    'User-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36'
}
from bs4 import BeautifulSoup
def function(a):
        r=requests.get(a,verify=False,headers=headers)
        r.encoding='utf-8'
        r=BeautifulSoup(r.text,'lxml') #以lxml為解碼器
        r=r.find('title')
        r=r.text
        r=r.strip()
        r1= re.compile(r'[<title> /]')
        r= r1.sub('' ,r )
        return r
tab['title'] = tab.apply(lambda x: function(x.hyperlink), axis = 1) 
```

```python
def function(a,b): 
    if b==0:
        r=requests.get(a,verify=False,headers=headers)
        r.encoding='utf-8'
        r=BeautifulSoup(r.text,'lxml') #以lxml為解碼器
        r=r.find_all('p')
        r='%s'*len(r) % tuple(r) 
        r=r.strip()
        r1= re.compile(r'[<p> /]')
        r= r1.sub('' ,r )
        return r
    else:
        r=requests.get(a,verify=False,headers=headers)
        r.encoding='utf-8'
        r=BeautifulSoup(r.text,'lxml') #以lxml為解碼器
        r=r.get_text() 
        r=r.splitsplit('(function (d, a, b, l, e, _)')[0].split('NOWnews關鍵字')[0].split('insertBefore(s, el)')[1].split('\n大')[1]
        r=r.strip()
        r1= re.compile(r'[<p> /]')
        r= r1.sub('' ,r )
        return r
#div,p分流
tab['result'] = tab.apply(lambda x: function(x.hyperlink,x.test), axis = 1) 
```
